## Convert Videos to WAV files for processing

```python
#Import relevant libraries
import os
import numpy as np
from moviepy.editor import *
from glob import glob
import re 
```

```python
# Identify all mp4s using glob falling under the target video folder and append them to a list
videos=glob('data/video/*.mp4')
#videos
#videos[0]
```

```python
videos
```

```python
#Uses MoviePy to convert mp4 video clips into required 

target_directory = './data/audio/'

for i,video in enumerate(videos):   
    if "lie" in video:
        main_clip = VideoFileClip(videos[i])
        audioclip = main_clip.audio 
        video_identifier = re.sub("(\D\D\D)","",video)
        file_name = str(target_directory + 'trial_lie_audio_' + str(video_identifier[0:3]) + '.wav')
        audioclip.write_audiofile(file_name,codec='pcm_s16le',fps=44100)
    elif "truth" in video:
        print(i,video)
        main_clip = VideoFileClip(videos[i])
        audioclip = main_clip.audio 
        video_identifier = re.sub("\D\D\D","",video)
        file_name = str(target_directory + 'trial_truth_audio_' + str(video_identifier[0:3]) + '.wav')
        audioclip.write_audiofile(file_name,codec='pcm_s16le',fps=44100)        
```
