## 7. Calculate player "power".

---

Define a function that will calculate a player's "power". Player power is defined as the sum of the base statistics all of their pokemon.

Your function will:

1. Accept the `players` dictionary, `pokedex` dictionary, and a player_id as arguments.
2. For the specified player_id, look up that player's pokemon and their level(s).
3. Find and aggregate the attack and defense values for each of the player's pokemon from the `pokedex` dictionary.
4. Print "[player name]'s power is [player power].", where the player power is the sum of the base statistics for all of their pokemon.
5. Return the player's power value.

Print out the pokemon power for each of your players.

````python
#1. This function takes the players database, pokemon_stats database and the index value of a player

def player_power(players, pokemon_stats, player_ident):
    #Takes players dictionary, pokemon stats dictionary and a player index value (player_ident) to 
    #and looks up the players pokemon (contained in pokemon_lookup) in the pokeman_stats dictionary
    #and returns the average of the attack, defense, special attack, and special defense points of ALL pokemon
    #associated with the player
    pokemon_lookup = []
    power_calculation = []
    
    #Loops through the nested keys in the players dictionary...
    for p_id, player in players.items():
        #...and if the inputed player_ident is equal to the key of the top level dictionary...
        if player_ident == p_id:
            #...loops through each of the players and appends the name of the pokemon to the pokemon_lookup list
            for ind_poke_player in player['player_pokemon']:
                pokemon_lookup.append(ind_poke_player)
    
    #loops through the nested keys in the pokemon_stats dictionary...
    for pokid, every_pokemon in pokemon_stats.items():
        #...and if the name of the pokemon is in the pokemon_lookup list...
        if every_pokemon['name'] in pokemon_lookup:
            #...appends the attack, defense, special attack and special defense values to the power_calculation variable
            power_calculation.append(every_pokemon['attack'])
            power_calculation.append(every_pokemon['defense'])
            #Not sure if special attack and special defense also needed to be added but added just in case
            power_calculation.append(every_pokemon['special attack']) 
            power_calculation.append(every_pokemon['special defense']) 
    
    #As some players may not have a pokemon, this first tries to calculate a power associated with a player
    #If the power_calculation returns an error (i.e. divisable by 0 due to no pokemon), then returns message
    try:
        power_calculation = sum(power_calculation)
        return print(players[player_ident]['player_name'],"'s power is", power_calculation)
    
    except:
        return print(players[player_ident]['player_name'],"doesn't have any pokemon")
````

````python


def full_pokedex(data):
    #takes data variable (which is the list of all pokemon) and 
    #transposes each column in data into new variables aligned with pokemon specs
    #and then iterates through each of the pokemon and appends them to a new poke_stats2 dictionary
    #returns this poke_stats2 dictionary

    #Note as original data contains duplicate IDs, the poke_dex_ids creates a unique ID list just based on range
    #of original data column index 0. 
    poke_dex_ids = list(range(len(list(list(zip(*data))[0]))))
    poke_names = list(list(zip(*data))[1])
    poke_types = list(list(zip(*data))[2])
    poke_totals = list(list(zip(*data))[3])
    poke_hps = list(list(zip(*data))[4])
    poke_atts = list(list(zip(*data))[5])
    poke_defs = list(list(zip(*data))[6])
    poke_spe_atts = list(list(zip(*data))[7])
    poke_spe_defs = list(list(zip(*data))[8])
    poke_speeds = list(list(zip(*data))[9])

    poke_stats2 = {}

    for poke_id in poke_dex_ids:
        poke_stats2[poke_dex_ids[poke_id]]= {'name': poke_names[poke_id],
                                           'type': poke_types[poke_id],
                                           'total': poke_totals[poke_id],
                                           'hp': poke_hps[poke_id],
                                           'attack': poke_atts[poke_id],
                                           'defense': poke_defs[poke_id],
                                           'special attack': poke_spe_atts[poke_id],
                                           'special defense': poke_spe_defs[poke_id],
                                           'speed': poke_speeds[poke_id]}
        
    return poke_stats2
````

### 9.1

What is the population mean and standard deviation of the "Total" attribute for all characters in the Pokedex?



````python
#1. Imports the math library to run the sqrt function below.
import math

#2. Initalises an empty total_append list to append values from total attribute
total_append = []

#3. Calls the full_pokedex function with the data set and iterates through the nested total key and appends the 
#   found value to the total_append variable
for a, b in full_pokedex(data).items():
    total_append.append(b['total'])

#4. Calcuates the total mean
total_mean = sum(total_append) / len(total_append)

#5. To calculate standard deviation starts with creating a empty list
total_std = []

#6. Iterates through all the elements in the original total_append and takes each value and takes from mean and squares it
#   and then appends this value to total_std list
for i in total_append:
    total_std.append((i - total_mean) ** 2)

#6. Takes the sqrt of the summed values in total_std divided by the number of values in total_std
total_std = math.sqrt(sum(total_std) / len(total_std))

#7. Prints to the console the mean and std. of Total attribute
print("The mean of 'Total' is", round(total_mean,2), "\nThe standard deviation is", round(total_std,2))
````
